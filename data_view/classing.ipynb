```python
import matplotlib.pyplot as plt

# 定义数据
labels1 = ['A', 'B', 'C']
sizes1 = [30, 40, 30]
colors1 = ['yellowgreen', 'gold', 'lightskyblue']

labels2 = ['D', 'E', 'F']
sizes2 = [20, 50, 30]
colors2 = ['pink', 'purple', 'coral']

# 绘制图表
fig, (ax1, ax2,ax3) = plt.subplots(1, 3)

ax1.set_title('Pie chart 1')
ax1.pie(sizes1, labels=labels1, colors=colors1, autopct='%1.1f%%')
ax1.axis('equal')

ax2.set_title('Pie chart 2')
ax2.pie(sizes2, labels=labels2, colors=colors2, autopct='%1.1f%%')
ax2.axis('equal')

ax3.set_title('Pie chart 3')
ax3.pie(sizes2, labels=labels2, colors=colors2, autopct='%1.1f%%')
ax3.axis('equal')

plt.show()
```
