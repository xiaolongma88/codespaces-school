```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 读取数据文件
data = pd.read_csv('./data_set/China-2019.csv')
```

```python
# 提取男女人口数据
male_data = data['M']
female_data = data['F']
```

```python
# 绘制金字塔

fig, ax = plt.subplots()
ax.set_xlabel('Population')
ax.set_ylabel('Age')
ax.set_title('Population Pyramid - 2019')
ax.barh(data['Age'], male_data, height=-0.8, color='blue')
ax.barh(data['Age'], -female_data, height=0.8, color='red')
```
