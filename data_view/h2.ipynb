```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv('./data_set/heart.csv')
```

```python
sns.boxplot(x='cp', y='chol', data=df)
plt.title('Cholesterol levels by chest pain type')
plt.xlabel('Chest pain type')
plt.ylabel('Cholesterol measurement (mg/dl)')
plt.show()
```

```python
sns.violinplot(x='cp', y='chol', data=df)
plt.title('Cholesterol levels by chest pain type')
plt.xlabel('Chest pain type')
plt.ylabel('Cholesterol measurement (mg/dl)')
plt.show()
```
